```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

```

```python
array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print(array_4x4)
```

```python
exponential_array = np.random.exponential(scale=1.0, size=(100000,))
uniform_array = np.random.rand(100000)
normal_array = np.random.normal(size=100000)
```

```python
plt.hist(exponential_array, bins=100, density=True, histtype='step', color='blue', label='Exponential')
plt.hist(normal_array, bins=100, density=True, histtype='step', color='green', label='Normal')
plt.hist(uniform_array, bins=100, density=True, histtype='step', color='red', label='Uniform')

plt.axis([0, 10, 0, 1.1])  # Adjust the x and y axis limits
plt.legend(loc='upper right')
plt.title('Random Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D plot of Z = X^2 + Y^2')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df.head()
```

```python
features = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
pearson_corr = features.corr(method='pearson')
spearman_corr = features.corr(method='spearman')

plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')
plt.show()


plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation Heatmap')
plt.show()
```
